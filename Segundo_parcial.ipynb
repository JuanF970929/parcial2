```python
import random
import matplotlib.pyplot as plt

# Lista de equipos de la liga Colombiana
equipos = ['Once Caldas', 'Deportivo Pereria', 'Cali', 'Pasto', 'Jaguares']

# GeneraciÃ³n aleatoria de partidos y goles
partidos = []
goles_local = []
goles_visitante = []
for i in range(len(equipos)):
  for j in range(len(equipos)):
    if i != j:
      partidos.append([equipos[i], equipos[j]])
      goles_local.append(random.randint(0,5))
      goles_visitante.append(random.randint(0,5))
      print("los partidos son: ", partidos)
      print("los goles de local son: ",goles_local)
      print("los goles de visitante son: " ,goles_visitante)

# 1. Cantidad de partidos por equipo
partidos_jugados = dict.fromkeys(equipos, 0)
for partido in partidos:
  partidos_jugados[partido[0]] += 1
  partidos_jugados[partido[1]] += 1
print("los partidos jugados son: ", partidos_jugados)

# 2. Cantidad de partidos ganados por equipo
partidos_ganados = dict.fromkeys(equipos, 0)
for i in range(len(partidos)):
  if goles_local[i] > goles_visitante[i]:
    partidos_ganados[partidos[i][0]] += 1
  elif goles_local[i] < goles_visitante[i]:
    partidos_ganados[partidos[i][1]] += 1
print("los partidos ganados son: ",partidos_ganados)
# 3. Cantidad de partidos perdidos por equipo
partidos_perdidos = dict.fromkeys(equipos, 0)
for i in range(len(partidos)):
  if goles_local[i] < goles_visitante[i]:
    partidos_perdidos[partidos[i][0]] += 1
  elif goles_local[i] > goles_visitante[i]:
    partidos_perdidos[partidos[i][1]] += 1
    print("los partidos perdidos son: ",partidos_perdidos)

# 4. Cantidad de partidos empatados por equipo
partidos_empatados = dict.fromkeys(equipos, 0)
for i in range(len(partidos)):
  if goles_local[i] == goles_visitante[i]:
    partidos_empatados[partidos[i][0]] += 1
    partidos_empatados[partidos[i][1]] += 1
print("los partidos empatados son: ",partidos_empatados)

# 5. Cantidad de goles de los equipos locales
goles_local_total = dict.fromkeys(equipos, 0)
for i in range(len(partidos)):
  goles_local_total[partidos[i][0]] += goles_local[i]
  print("los goles de local son: ",goles_local)

plt.bar(goles_local_total.keys(), goles_local_total.values())
plt.xticks(rotation=90)
plt.title('Goles de los equipos locales')
plt.show()
print("total de goles de local: " ,goles_local_total)

# 6. Cantidad de goles de los equipos visitantes
goles_visitante_total = dict.fromkeys(equipos, 0)
for i in range(len(partidos)):
  goles_visitante_total[partidos[i][1]] += goles_visitante[i]
print(goles_visitante_total)
plt.bar(goles_visitante_total.keys(), goles_visitante_total.values())
plt.xticks(rotation=90)
plt.title('Goles de los equipos visitantes')
plt.show()

# 7. Cantidad total de goles de todos los partidos
goles_total = sum(goles_local) + sum(goles_visitante)
print("goles totales: ",goles_total)
```
